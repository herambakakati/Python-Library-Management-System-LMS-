# Library Management System(LMS)

```python
### --- Step 1: Login System ---
def login_form(display_welcome = True):
    if display_welcome:
        print("\nWelcome to the library Management System (LMS)\n")
                
    credentials = {"admin": "123", "librarian": "123"}
    attempts = 3
        
    while attempts > 0:
        while True:
            uid = input("Enter your username:  ").strip().lower()
            if uid == "":
                print("Error: Username cannot be blank!")
                continue
            else:
                break

        while True:
            password = input("Enter your password: ").strip()
            if password == "":
                print("Error: Password cannot be blank!")
                continue
            else:
                break
                           
        if uid in credentials and credentials[uid] == password:
            print(f"\nLogin successful! Welcome {uid.title()}.")
            main_menu(books)
            return
            
        else:
            attempts -= 1
            print(f"\nInvalid credentials! Attempts remaining: {attempts}\n" )
                
        if attempts == 0:
            print("Access denied! Too many failed login attempts.")
            exit()    
            
            
#--- Step 2: Initial Book List ---
books = [
    {"title": "Assam History", "author": "Padman Nath Gohain Baruah", "available": True,"issued_to": ""},
    {"title": "The Discovery of India", "author": "Jawaharlal Nehru", "available": True,"issued_to": ""},
    {"title": "Wings of Fire", "author": "A.P.J. Abdul Kalam", "available": False,"issued_to": "Hera"},
    {"title": "The Guide", "author": "R.K. Narayan", "available": True,"issued_to": ""},
    {"title": "Train to Pakistan", "author": "Khushwant Singh", "available": True,"issued_to": ""},
    {"title": "God of Small Things", "author": "Arundhati Roy", "available": False,"issued_to": "Queen"},
    {"title": "The White Tiger", "author": "Aravind Adiga", "available": True,"issued_to": ""},
    {"title": "Midnight’s Children", "author": "Salman Rushdie", "available": True,"issued_to": ""},
    {"title": "The Immortals of Meluha", "author": "Amish Tripathi", "available": False,"issued_to": "Kakati"},
    {"title": "The Palace of Illusions", "author": "Chitra Banerjee Divakaruni", "available": True,"issued_to": ""},
    {"title": "Five Point Someone", "author": "Chetan Bhagat", "available": True,"issued_to": ""},
    {"title": "To Kill a Mockingbird", "author": "Harper Lee", "available": False,"issued_to": "Devi"}
]


# --- Step 3: Display Books Function ---
def display_book(book_list):
    print("\nAvailable books in the library:\n")
    print(f"{'Title':<30}{'Author':<35}{'Status':<20}")
    print("_"*80)
    
    for i in book_list:
        status = "Available" if i["available"] else f"issued to {i['issued_to']}"
        print(f"{i['title'].title():<30}{i['author'].title():<35}{status:<20}")
    print()
    
    
# --- Step 4: Add Book Function ---
def add_book(book_list):
    print("\nAdd New Book")
    print("-"*25)
    
    while True:
        while True:
            title = input("Enter Name of the book: ").strip().lower()
            if title == "":
                print("Error: Name of the book cannot be blank!\n")
                continue
            else:
                break
        while True:
            author = input("Enter the author name: ").strip().lower()
            if author == "":
                print("Error: Author name cannot be blank!")
                continue
            else:
                break

        for i in book_list:
            if i["title"].lower() == title.lower():
                print(f"'{title.title()}' already exists in the library!\n")
                break
                         
        else:
            new_book = {"title": title, "author": author,"available":True,"issued_to":""}
            book_list.append(new_book)
            print(f"\n'{title.title()}' by '{author.title()}' has been added  successfully!")
    
        while True:
            choice = input("\nWould you like to continue adding more books? (Yes/No): ").lower().strip()
            
            if choice == "":
                print("Error: choice cannot be blank!\n")
                continue
                
            elif choice == "no":
                print("The book list has been updated successfully.")
                display_book(book_list)
                return
            
            elif choice == "yes":
                break
            else:
                print("Error: Please enter 'Yes' or 'No' only! ")
        
           
# --- Step 5: Issue Book Function ---
def issue_book(book_list):
    print("\nIssue Book")
    print("-"*25)

    while True:
        title = input("Enter the book name you want to issue: ").strip()
        if title == "":
            print("Error: Book name cannot be blank!\n")
            continue

        found = False

        for j in book_list:
            if j["title"].lower() == title.lower():
                found = True
                if j["available"]:
                    while True:
                        student = input("Please enter the name of the student who will receive the book: ").strip()
                        if student == "":
                            print("Error: Student name cannot be blank!\n")
                        else:
                            break
                    j["available"] = False
                    j["issued_to"] = student
                    print(f"'{j['title'].title()}' has been issued to {student.title()}.")
                else:
                    print(f"Sorry, '{j['title'].title()}' is already issued to '{j['issued_to'].title()}'.")
                break
        if not found:
            print(f"Sorry, the book '{title.title()}' is not found in the library.")

        while True:
            issue = input("\nWould you like to issue another book? (Yes/No) : ").lower().strip()
            if issue == "":
                print("Error: choice cannot be blank!")
                continue
            
            elif issue == "no":
                print("\nUpdated list of available books: ")
                display_book(book_list)
                return
            
            elif issue == "yes":
                break
            else:
                print("Error: Please enter 'Yes' or 'No' only! \n ")
        

# --- Step 6: Return Book Function ---
def return_book(book_list):
    print("\nReturn Book")
    print("-"*25)

    while True:
        title = input("\nEnter the name of the book to return: ").strip()
        if title == "":
            print("Error: Book name cannot be blank!")
            continue
            
        found = False
        
        for k in book_list:
            if k["title"].lower() == title.lower():
                found = True
                if not k["available"]:
                    print(f"'{k['title'].title()}' was issued to '{k['issued_to']}'.")
                    
                    while True:
                        confirm = input("Do you want to confirm the return? (Yes/No):").strip().lower()
                        
                        if confirm == "":
                            print("Error: Confirmation cannot be blank!")
                            continue
                            
                        elif confirm == "yes":
                            k["available"] = True
                            k["issued_to"] = ""
                            print(f"'{k['title'].title()}' has been successfully returned!")
                            break
                        
                        elif confirm == "no":
                            print("Book return has been cancelled.")
                            break
                        
                        else: 
                             print("Error: Please enter 'Yes' or 'No' only!")
                else:
                    print(f"'{k['title'].title()}' is already available in the library. ")
                break
                
        if not found:
            print(f"Sorry, '{title.title()}' is not found in the library.")
            
        while True:
            book_return = input("\nWould you like to return another book? (Yes/No) : ").lower().strip()
            if book_return == "":
                print("Error: choice cannot be blank!")
                continue
                
            elif book_return == "no":
                print("\nUpdated list of available books:")
                display_book(book_list)
                return
            
            elif book_return == "yes":
                break
            else:
                print("Error: Please enter 'Yes' or 'No' only!")

# --- Step 7: Main Menu ---
def main_menu(book_list):
    
    while True:
        print("\nLibrary Menu:")
        print("\t1. Display all books")
        print("\t2. Add a book")
        print("\t3. Issue a book")
        print("\t4. Return a book")
        print("\t5. Exit")

        choice = input("\nSelect your choice: ").strip()

        if choice == "":
            print("Error: choice cannot be blank!")

        elif choice == "1":
            display_book(book_list)

        elif choice == "2":
            add_book(book_list)

        elif choice == "3":
            issue_book(book_list)

        elif choice == "4":
            return_book(book_list)

        elif choice == "5":
            print("\nThank you for using the Library Management System. Have a great day!")
            
            while True:
                again = input("\nWould you like to log in again? (Yes/No): ").lower().strip()

                if again =="":
                    print("Error: choice cannot be blank!")
                    continue

                elif again == "yes":
                    login_form(display_welcome=False)
                    return
                
                elif again == "no":
                    print("\nYour session has ended. Thank you for visiting!\n")
                    return

                else:
                    print("Error: Please enter 'Yes' or 'No' only!")
                    continue
                          
        else:
            print("Error: Invalid choice. Please enter a number between 1 and 5 !")
            continue

# --- Step 8: Run Program ---

if __name__ == "__main__":
    login_form(display_welcome=True)
```
